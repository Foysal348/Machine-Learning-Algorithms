```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
#Generate data
np.random.seed(0)
data_1=np.random.uniform(10,50,100)
data_2=np.random.uniform(10,40,100)
data=np.column_stack((data_1,data_2))

#Normalization
model=MinMaxScaler()
scaled_data=model.fit_transform(data)
scaled_data_1=scaled_data[:,0]
scaled_data_2=scaled_data[:,1]

fig,axes=plt.subplots(3,2,figsize=(17,8))
#1st plot 
axes[0,0].scatter(data_1,data_2)
axes[0,0].set_title('Before Normalization')
axes[0,0].set_xlabel('Feature A')
axes[0,0].set_ylabel('Feature B')
#2nd Plot
axes[0,1].scatter(scaled_data_1,scaled_data_2)
axes[0,1].set_title('After Normalization')
axes[0,1].set_xlabel('Scaled_Feature A')
axes[0,1].set_ylabel('Scaled_Feature B')

#Histogram
#1st plot
sns.histplot(data_1,label='Feature A',alpha=0.4,bins=10,ax=axes[1,0],color='skyblue',kde=True)
sns.histplot(data_2,label='Feature B',alpha=0.4,bins=10,ax=axes[1,0],color='r',kde=True)
axes[1,0].set_title('Before Normalization')
#2nd Plot
sns.histplot(scaled_data_1,label='Scaled_Feature A',alpha=0.4,bins=10,ax=axes[1,1],color='skyblue',kde=True)
sns.histplot(scaled_data_2,label='Scaled_Feature B',alpha=0.4,bins=10,ax=axes[1,1],color='r',kde=True)
axes[1,1].set_title('After Normalization')

#Ploting
axes[2,0].bar(data_1,data_2)

axes[2,1].bar(data_1,data_2,bins=10)
plt.tight_layout()
axes[1,0].legend()
plt.show()
```
