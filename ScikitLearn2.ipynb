```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
np.random.seed(50)
num_samples=500
years_of_experience=np.random.randint(2,21,size=num_samples)
slope=(200_000-60_000)/18
intercept=60_000
salaries=slope*years_of_experience+intercept+np.random.normal(0,10_000)
data={'Years_of_Experience':years_of_experience,'Salary':salaries}
df=pd.DataFrame(data)
df.describe()
plt.figure(figsize=(10,6))
sns.scatterplot(x='Years_of_Experience', y='Salary', data=df, color='red')
sns.regplot(x='Years_of_Experience', y='Salary', data=df, color='blue')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title('Linear Regression Salary')
plt.show()
```
